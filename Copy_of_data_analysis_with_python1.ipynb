```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file using pandas
data = pd.read_csv('coursea_data.csv')

# Filter the data for courses with a rating greater than 4.7
high_rated_courses = data[data['course_rating'] > 4.7]

high_rated_courses
# Count the number of highly rated courses enrolled based on 'course_Certificate_type'
count_by_certificate_type = high_rated_courses['course_Certificate_type'].value_counts()

# Plotting the count of highly rated courses enrolled based on certificate type
plt.figure(figsize=(8, 6))  # Define the figure size
count_by_certificate_type.plot(kind='bar', color='skyblue')  # Bar plot
plt.title('Number of Highly Rated Courses Enrolled by Certificate Type')
plt.xlabel('Certificate Type')
plt.ylabel('Number of Courses')
plt.grid(axis='y')  # Show grid on the y-axis
plt.show()  # Display the plot


```

```python
import pandas as pd
filename='medical_examination.csv'
df=pd.read_csv(filename)
```

```python
df.info()
```

```python
df['height_in_meters']=df['height']/100

df.info()
```

```python
df
```

```python
df['BMI']=df['weight']/(df['height_in_meters']**2)
df
```

```python
df['overweight']=(df['BMI']>25).astype(int)
df
```

```python
df['cholesterol'] = (df['cholesterol'] > 1).astype(int)
df
```

```python
df['gluc'] = (df['gluc'] > 1).astype(int)
df
```

```python
df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# Melt the DataFrame into long format
df_long = pd.melt(df, id_vars=['age', 'sex', 'height', 'weight', 'gluc', 'cholesterol', 'overweight', 'smoke'], value_vars=['cholesterol', 'gluc', 'overweight'])

# Check if the DataFrame is empty
if df_long.empty:
    print("The DataFrame is empty. Cannot create a catplot.")
else:
    # Create a catplot using seaborn
    g = sns.catplot(x='variable', hue='value', col='Cardio', data=df_long, kind='count')

    # Set axis labels and plot title
    g.set_axis_labels('Categorical Features', 'Count')
    g.set_titles('Cardio = {col_name}')

    # Show the plot
    plt.show()
```
