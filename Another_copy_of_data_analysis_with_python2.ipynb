```python
import pandas as pd
filename=('medical_examination.csv')
df=pd.read_csv(filename)
```

```python
df
```

```python
# Assuming your DataFrame is named 'df'

# Filter out incorrect diastolic and systolic pressure data
df = df[df['ap_lo'] <= df['ap_hi']]

# Filter out incorrect height data
height_percentiles = df['height'].quantile([0.025, 0.975])
df = df[(df['height'] >= height_percentiles.loc[0.025]) & (df['height'] <= height_percentiles.loc[0.975])]

# Filter out incorrect weight data
weight_percentiles = df['weight'].quantile([0.025, 0.975])
df = df[(df['weight'] >= weight_percentiles.loc[0.025]) & (df['weight'] <= weight_percentiles.loc[0.975])]

# Print or display the cleaned DataFrame
print(df)
```

```python
df.corr()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
correlation_matrix = df.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(10, 8))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlation_matrix, mask=mask, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5, ax=ax)

# Set plot title
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def draw_bar_plot():
   df = pd.read_csv('fcc-forum-pageviews.csv')
        # Assuming your dataset has columns named 'date' and 'value'
   df['date'] = pd.to_datetime(df['date'])

   df['year'] = df['date'].dt.year
   df['month'] = df['date'].dt.month_name()

    # Group by year and month, calculate the mean of page views
   df_grouped = df.groupby(['year', 'month'], as_index=False)['value'].mean()

    # Pivot the DataFrame to have months as columns
   df_pivot = df_grouped.pivot(index='year', columns='month', values='value')

    # Reorder columns to have months in chronological order
   month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
   df_pivot = df_pivot.reindex(month_order, axis=1)

    # Plotting
   plt.figure(figsize=(10, 6))
   df_pivot.plot(kind='bar', stacked=True)

   plt.title('Average Daily FCC PAGEVIEWS for Each Month Grouped by Year')
   plt.xlabel('Years')
   plt.ylabel('Average Value')
   plt.legend(title='Months', bbox_to_anchor=(1.05, 1), loc='upper left')
   plt.show()

# Call the function to draw the bar plot
draw_bar_plot()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def draw_box_plot():
    df = pd.read_csv('fcc-forum-pageviews.csv')

    # Assuming your dataset has columns named 'date' and 'value'
    df['date'] = pd.to_datetime(df['date'])

    # Extract year and month from the date
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month_name()

    # Prepare data for year-wise and month-wise box plots
    df_year = df.copy()
    df_month = df.copy()

    # Plotting
    plt.figure(figsize=(15, 6))

    # Year-wise Box Plot (Trend)
    plt.subplot(1, 2, 1)
    sns.boxplot(x='year', y='value', data=df_year)
    plt.title('Year-wise Box Plot (Trend)')
    plt.xlabel('Year')
    plt.ylabel('Page Views')

    # Month-wise Box Plot (Seasonality)
    plt.subplot(1, 2, 2)
    sns.boxplot(x='month', y='value', data=df_month, order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
    plt.title('Month-wise Box Plot (Seasonality)')
    plt.xlabel('Month')
    plt.ylabel('Page Views')

    plt.tight_layout()
    plt.show()

# Call the function to draw the box plots
draw_box_plot()
```
